# Тема 7 “Ряды”

1. Исследовать ряд на сходимость, используя признак д’Аламбера:

```python
%%latex
$$\sum_{n=1}^{\infty}\frac{n^n}{(n!)^2}$$
$$\lim_{n\to\infty}\left|\frac{\frac{(n+1)^{n+1}}{((n+1)!)^2}}{\frac{n^n}{(n!)^2}}\right|=\lim_{n\to\infty}
\frac{(n+1)^n(n+1)(n!)^2}{(n!)^2(n+1)^2n^n}=\lim_{n\to\infty}\frac{(n+1)^n}{n^n(n+1)}=\lim_{n\to\infty}
\frac{(1+\frac{1}{n})^n}{n+1}=\frac{1}{\infty}=0<1\text{ - ряд сходится}$$
```

2. Исследовать ряд на сходимость, используя радикальный признак Коши:

```python
%%latex
$$\sum_{n=1}^{\infty}\frac{n}{2^n}$$
$$\lim_{n\to\infty}\sqrt[n]{\frac{n}{2^n}}=\lim_{n\to\infty}\frac{\sqrt[n]{n}}{2}=\frac{1}{2}<1\text{ - ряд сходится}$$
```

3. Исследовать ряд на сходимость, используя признак Лейбница:

```python
%%latex
$$\sum_{n=1}^{\infty}\frac{(-1)^n}{n+\ln{n}}$$
$$1.\quad \frac{1}{n+\ln n}>\frac{1}{(n+1)+\ln(n+1)}$$
$$2.\quad\lim_{n\to\infty}\frac{1}{n+\ln n}=0$$$$$$
$$\text{Ряд сходится абсолютно}$$
```

4. Исследовать ряд на сходимость, используя признак Раабе:

```python
%%latex
$$\sum_{n=1}^{\infty}\frac{3^n}{2^n}$$
$$\lim_{n\to\infty}n\left(\frac{\frac{3^n}{2^n}}{\frac{3^{n+1}}{2^{n+1}}}-1\right)=
\lim_{n\to\infty}n\left(\frac{3^n*2^{n+1}}{2^n*3^{n+1}}-1\right)=
\lim_{n\to\infty}n\left(\frac{2}{3}-1\right)=-\infty\text{ - ряд расходится}$$
```

5. Разложить функцию по Тейлору в единице:

```python
%%latex
$$f(x)=ln(16x^2)$$$$$$
$$f'(1)=\frac{1}{16x^2}*32x=\frac{2}{x}=2;\quad f''(1)=-\frac{2}{x^2}=-2;\quad f^{(3)}(1)=\frac{4}{x^3}=4;\quad
f^{(4)}(1)=-\frac{12}{x^4}=-12;\quad f^{(5)}(1)=\frac{48}{x^5}=48$$$$$$
$$\ln(16x^2)=\ln(16)+2(x-1)+\frac{-2}{2!}(x-1)^2+\frac{4}{3!}(x-1)^3+\frac{-12}{4!}(x-1)^4+\frac{48}{5!}(x-1)^5+...
=\ln(16)+\sum_{n=1}^{\infty}\frac{(-1)^{n-1}2(n-1)!}{n!}(x-1)^n$$
$$\ln(16x^2)=\ln(16)+\sum_{n=1}^{\infty}-\frac{2(-1)^n}{n}(x-1)^n$$
```

6. Дана функция f(x)=x^2
a. Разложить функцию в ряд Фурье по косинусам на отрезке [-2; 0].
b. Построить график функции и ее разложения.

```python
%%latex
$$T=2;\quad b_n=0\text{ - т.к. функция четная}$$
$$a_0=\frac{2}{2}\int_{-2}^{0}x^2dx=\left.\frac{x^3}{3}\right|^0_{-2}=0-\frac{-8}{3}=\frac{8}{3}$$
$$\left.a_n=\frac{2}{2}\int_{-2}^{0}x^2\cos(\frac{\pi nx}{2})dx=\left(2x^2\frac{\sin(\pi n \frac{x}{2})}{\pi n}+
8\frac{\cos(\pi n \frac{x}{2})}{\pi^2n^2}-16\frac{\sin(\pi n \frac{x}{2})}{\pi^3n^3}\right)\right|_{-2}^0=$$
$$=(0-8\frac{\sin(-\pi n)}{\pi n})+(0+16\frac{\cos(-\pi n)}{\pi^2 n^2})-(0-16\frac{\sin(-\pi n)}{\pi^3 n^3})=
16\frac{\cos(\pi n)}{\pi^2 n^2}$$
$$f(x)=\frac{a_0}{2}+\sum_{n=1}^{\infty}a_n\cos(\frac{\pi nx}{T})=
\frac{4}{3}+\frac{16}{\pi^2}\sum_{n=1}^{\infty}\frac{(-1)^n}{n^2}\cos\left(\pi n\frac{x}{2}\right)$$
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

N_max = 5
x = np.linspace(-2, 0, 101)

fur = 0
for n in range(1, N_max + 1):
    fur += (-1)**n*np.cos(np.pi*n*x/2)/n**2
y = fur*16/np.pi**2+4/3

plt.plot(x, x**2)
plt.plot(x, y)
plt.legend(labels=['x**2','Fourier'])
plt.grid();
```
